Boxplot COM outliers - Cart√µes Amarelos Time da Casa

```python
import pandas as pd           # Para manipula√ß√£o do dataset
import matplotlib.pyplot as plt  # Para cria√ß√£o dos gr√°ficos


# 1. Carregar o dataset
caminho = "mydata.csv"        # Caminho do arquivo CSV
df = pd.read_csv(caminho)     # L√™ o arquivo e cria um DataFrame (tabela)


# 2. Selecionar a coluna de interesse
dados = df["home_yellow"]


# 3. Criar o Boxplot COM mostrar outliers
plt.figure(figsize=(6, 4))                  # Define tamanho do gr√°fico
plt.boxplot(dados, showfliers=True)         # "showfliers=True" mostra os outliers (padr√£o)
plt.title("Boxplot - Home Yellow (com outliers)")  # T√≠tulo do gr√°fico
plt.ylabel("Cart√µes amarelos")              # R√≥tulo do eixo Y
plt.grid(True, linestyle="--", alpha=0.6)   # Adiciona grade para facilitar a leitura
plt.show()                                  # Exibe o gr√°fico
```
