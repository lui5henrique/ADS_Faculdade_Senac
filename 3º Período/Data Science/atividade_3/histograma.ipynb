Histograma - Cart√µes Amarelos do Time da Casa

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mydata.csv')
print(df['home_yellow'].head(5))

plt.hist(df['home_yellow'], bins=range(df['home_yellow'].min(), df['home_yellow'].max() + 2), color = 'yellow', edgecolor = 'gray')
plt.title('Histograma - Cart√µes Amarelos Time da Casa')
plt.xlabel('N√∫mero de Cart√µes Amarelos')
plt.ylabel('Frequ√™ncia')
plt.grid(axis='y', alpha=0.25)

plt.show()
```
